## 컬럼별 전처리

```python
import re

# title 컬럼에서 질문\n, date 컬럼에서 직성일\n 제거
df['title'] = df['title'].str.replace(r'^질문\n', '', regex=True)
df['date'] = df['date'].str.replace(r'^작성일\n', '', regex=True)
df['question'] = df['question'].str.replace(r'안녕하세요', '', regex=True)
df['answer'] = df['answer'].str.replace(r'안녕하세요', '', regex=True)
df['answer'] = df['answer'].apply(lambda x: re.sub(r"대한의사협회 네이버 지식 상담의사 \d{3,} 입니다\.?", "", x).strip())
```

## 문장으로 분리

```python
import pandas as pd
import numpy as np
import kss

# 문장 분리 함수
def sentences(text):
    if isinstance(text, str):  # 텍스트가 문자열인지 확인
        sentences = kss.split_sentences(text)
        return sentences
    return np.nan  # 비문자형 값이나 NaN이면 NaN 반환

# 'title'과 'question' 컬럼을 합치고 문장 분리 함수 적용
df['sentences'] = (df['title'] + ' ' + df['question']).apply(sentences)
```

##증상사전으로 문장 추출

```python
symptoms_df=pd.read_csv('/content/drive/MyDrive/24-2 Capston/질환별 증상 조합사전.csv')
symptoms_df.head()
```

```python
symptoms=[]
for i in range(len(symptoms_df)):
  match=symptoms_df['단어 조합'][i]
  match=re.findall(r"\('(.+?)', '(.+?)'\)", match)
  word1,word2=match[0]
  symptoms.append((word1, word2))
symptoms[:5]
```

```python
from tqdm import tqdm

# 문장에서 두 단어가 모두 있는 문장만 남기는 함수
def filter_sentences(text):
    for symptom_pair in symptoms:
        word1, word2 = symptom_pair
        if word1 in text and word2 in text:  # 두 단어가 모두 포함된 경우
            return text
    return ''  # 조건에 맞는 문장이 없으면 빈 문자열 반환

filtered = []

for text in tqdm(df['sentences'], desc="Filtering sentences"):
    filtered_per_post = []
    if isinstance(text, list):  # text가 리스트인 경우에만 반복
        for t in text:
            filtered_text = filter_sentences(t)
            if filtered_text:  # 빈 문자열이 아닌 경우에만 추가
                filtered_per_post.append(filtered_text)
    else:
        filtered_per_post.append(' ')
    filtered.append(filtered_per_post)  # 빈 리스트는 NaN으로 저장

# 리스트를 문자열로 결합
df['filtered_sentences'] = filtered
df['filtered_sentences'] = df['filtered_sentences'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
```

##특수문자 제거

```python
import re

# 특수문자 제거 함수 (온점 제외)
def remove_special_characters(text):
    if isinstance(text, list):  # 리스트 형태의 경우, 각 문장에 적용
        return [re.sub(r'[^가-힣a-zA-Z0-9\s.]', '', sentence) for sentence in text]
    return re.sub(r'[^가-힣a-zA-Z0-9\s.]', '', text)

# content 컬럼의 특수문자 제거 (온점 제외) 적용
df['filtered_sentences'] = df['filtered_sentences'].apply(remove_special_characters)
df.head()
```
