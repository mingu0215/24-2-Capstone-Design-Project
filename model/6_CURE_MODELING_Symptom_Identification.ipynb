```python
#evidence 관련 함수
# 토크나이징 (input_ids 및 attention_mask 생성)
def create_attention_mask(input_ids):
    return [1] * len(input_ids) + [0] * (max_length - len(input_ids))

# factor 증거(evidence) 결합
def combine_factor_inputs(evidence_input_ids_list, evidence_attention_mask_list):
    factor_input_ids, factor_attention_mask = [], []
    total_length = 0
    for evidence_input_ids, evidence_attention_mask in zip(evidence_input_ids_list, evidence_attention_mask_list):
        truncated_input_ids = evidence_input_ids[:max_length]
        truncated_attention_mask = evidence_attention_mask[:max_length]
        remaining_length = max_length - total_length
        if remaining_length <= 0:
            break
        factor_input_ids.extend(truncated_input_ids[:remaining_length])
        factor_attention_mask.extend(truncated_attention_mask[:remaining_length])
        total_length += len(truncated_input_ids)
    # 길이를 max_length로 제한
    return factor_input_ids[:max_length], factor_attention_mask[:max_length]
```

```python
#evidence
all_factor_input_ids = []
all_factor_attention_mask = []

for _, row in df.iterrows():  # iterrows()로 각 행 반복
    cause_evidence = row['cause_evidence']
    freq_evidence = row['freq_evidence']
    duration_evidence = row['duration_evidence']
    age_evidence = row['age_evidence']
    social_evidence = row['social_evidence']
    academic_evidence = row['academic_evidence']
    occupational_evidence = row['occupational_evidence']
    life_threatening_evidence = row['life-threatening_evidence']

    # evidence들을 각 토큰화된 input_ids로 변환
    evidence_fields = [
        cause_evidence, freq_evidence, duration_evidence, age_evidence,
        social_evidence, academic_evidence, occupational_evidence, life_threatening_evidence
    ]

    evidence_input_ids_list = [
        tokenizer.encode(evidence, add_special_tokens=True, truncation=True,
                         padding='max_length', max_length=max_length)
        for evidence in evidence_fields
    ]

    # attention_mask 생성
    evidence_attention_mask_list = [create_attention_mask(input_ids) for input_ids in evidence_input_ids_list]

    # factor 증거(evidence) 결합
    factor_input_ids, factor_attention_mask = combine_factor_inputs(
        evidence_input_ids_list, evidence_attention_mask_list
    )

    # 누적
    all_factor_input_ids.append(factor_input_ids)
    all_factor_attention_mask.append(factor_attention_mask)

# 최종 텐서로 변환
factor_input_ids = torch.tensor(all_factor_input_ids)
factor_attention_mask = torch.tensor(all_factor_attention_mask)

print(factor_input_ids.shape)  # (5926, 128)
print(factor_attention_mask.shape)  # (5926, 128)
```

## Modeling

### BERT symptom(Symptom Identification)

```python
# SpectralNorm 클래스
class SpectralNorm(nn.Module):
    def __init__(self, module, name='weight', power_iterations=1, eps=1e-12):
        super(SpectralNorm, self).__init__()
        self.module = module
        self.name = name
        self.power_iterations = power_iterations
        self.eps = eps
        self.register_buffer('u', torch.randn(1, module.weight.size(0)))
        self.register_buffer('v', torch.randn(1, module.weight.size(1)))

    def forward(self, *args):
        weight = self.module.weight
        for _ in range(self.power_iterations):
            self.u = F.normalize(torch.matmul(self.u, weight), dim=-1, eps=self.eps)
            self.v = F.normalize(torch.matmul(self.v, weight.T), dim=-1, eps=self.eps)
        sigma = torch.matmul(self.u, torch.matmul(weight, self.v.T))
        weight = weight / sigma
        return self.module(*args)

# SNGP 관련 클래스 및 함수들
def BertLinear(i_dim, o_dim, bias=True):
    m = nn.Linear(i_dim, o_dim, bias)
    nn.init.normal_(m.weight, std=0.02)
    if bias:
        nn.init.constant_(m.bias, 0.)
    return m

def RandomFeatureLinear(i_dim, o_dim, bias=True, require_grad=False):
    m = nn.Linear(i_dim, o_dim, bias)
    nn.init.normal_(m.weight, mean=0.0, std=0.05)
    m.weight.requires_grad = require_grad
    if bias:
        nn.init.uniform_(m.bias, a=0.0, b=2. * math.pi)
        m.bias.requires_grad = require_grad
    return m

class SNGP(nn.Module):
    def __init__(self, backbone, hidden_size=256, num_labels=48, gp_kernel_scale=2.0, num_inducing=128,
                 gp_output_bias=0., n_power_iterations=1, spec_norm_bound=0.95, device='cuda'):
        super(SNGP, self).__init__()
        self.backbone = backbone
        self.num_labels = num_labels
        self.hidden_size = hidden_size
        self.gp_kernel_scale = gp_kernel_scale
        self.num_inducing = num_inducing

        # BERT의 마지막 레이어
        self.last_pooled_layer = SpectralNorm(BertLinear(hidden_size, hidden_size),
                                               power_iterations=n_power_iterations)

        self.gp_input_scale = 1. / math.sqrt(gp_kernel_scale)
        self._random_feature = RandomFeatureLinear(hidden_size, num_inducing)
        self._gp_output_layer = nn.Linear(num_inducing, num_labels, bias=False)
        self._gp_output_bias = torch.tensor([gp_output_bias] * num_labels).to(device)

        self._gp_input_normalize_layer = torch.nn.LayerNorm(hidden_size)
        self.precision_matrix = torch.nn.Parameter(torch.eye(num_inducing).to(device), requires_grad=False)

    def gp_layer(self, gp_inputs):
        # gp_inputs를 정규화
        gp_inputs = self._gp_input_normalize_layer(gp_inputs)

        # 랜덤 특징 변환
        gp_feature = self._random_feature(gp_inputs)

        # cos() 함수로 변환 후 스케일링
        gp_feature = torch.cos(gp_feature) * self.gp_input_scale
        gp_output = self._gp_output_layer(gp_feature) + self._gp_output_bias

        return gp_feature, gp_output  # 두 값을 모두 반환

    def forward(self, input_ids, attention_mask):
        # BERT 모델에서 나온 출력
        outputs = self.backbone(input_ids=input_ids, attention_mask=attention_mask)

        # 'last_hidden_state'에서 첫 번째 토큰(CLS)을 추출
        pooled_output = outputs.last_hidden_state[:, 0, :]  # CLS 토큰 추출 (배치 크기, hidden_size)

        # SNGP 레이어에 입력
        gp_feature, logits = self.gp_layer(pooled_output)

        return logits  # logits만 반환


# BERT + SNGP 결합 모델
class BertMultiLabelClassificationWithSNGP(nn.Module):
    def __init__(self, model_name, num_labels, dropout_ratio=0.2, num_samples=10):
        super(BertMultiLabelClassificationWithSNGP, self).__init__()
        self.bert = AutoModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(dropout_ratio)
        self.num_samples = num_samples
        self.sngp = SNGP(self.bert, hidden_size=self.bert.config.hidden_size, num_labels=num_labels)
        # self.fc의 입력 차원을 96에서 49로 수정
        self.fc = nn.Linear(96, num_labels)

    def forward(self, input_ids, attention_mask, symptom_uncertainty=None):
        self.train()
        predictions, uncertainties = [], []

        for _ in range(self.num_samples):
            with autocast():  # 모델의 계산을 반정밀도에서 진행
                bert_outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
                pooled_output = bert_outputs.last_hidden_state[:, 0, :]
                pooled_output = self.dropout(pooled_output)

                gp_feature, gp_output = self.sngp.gp_layer(pooled_output)
                probabilities = torch.sigmoid(gp_output)

                predictions.append(probabilities)
                uncertainties.append(torch.std(probabilities, dim=0))

        mean_predictions = torch.mean(torch.stack(predictions), dim=0)  # [32, 48]
        mean_uncertainty = torch.mean(torch.stack(uncertainties), dim=0)  # [48]

        # mean_uncertainty를 [32, 1]로 확장
        mean_uncertainty = mean_uncertainty.unsqueeze(0)  # [1, 48]
        mean_uncertainty = mean_uncertainty.expand(mean_predictions.size(0), -1)  # [32, 48]으로 확장

        # symptom_uncertainty가 None이 아닐 때 combined_uncertainty를 만듦
        if symptom_uncertainty is not None:
            combined_uncertainty = torch.cat([symptom_uncertainty.unsqueeze(1), mean_uncertainty], dim=1)  # [32, 49] 형태
        else:
            combined_uncertainty = mean_uncertainty  # [32, 48]

        # mean_predictions와 combined_uncertainty를 결합
        combined_input = torch.cat([mean_predictions, combined_uncertainty], dim=1)  # [32, 96]로 결합

        # fc 계층을 통과
        logits = self.fc(combined_input)  # [32, 49] 형태로 처리됨

        return logits, mean_uncertainty  # logits과 mean_uncertainty 반환
```

```python
dataset = TensorDataset(
    text_input_ids,
    text_attention_mask,
    torch.tensor(symptom_labels, dtype=torch.float32)
)

# 훈련 세트와 검증 세트 분할
train_size = int(0.9 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(dataset, [train_size, val_size])

# 데이터 로더 생성
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
```

```python
# 모델, 옵티마이저, 스케줄러, 손실 함수 초기화
bert_symp = BertMultiLabelClassificationWithSNGP('klue/bert-base', num_labels=48, dropout_ratio=0.2).to(device)
optimizer = AdamW(bert_symp.parameters(), lr=3e-5)
total_steps = len(train_loader) * 20  # 총 훈련 스텝 계산
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)
loss_func = BCEWithLogitsLoss().to(device)

# 훈련 및 검증 손실 기록
train_losses = []
val_losses = []

# 훈련 루프
bert_symp.train()
early_stop_counter = 0
last_loss = float('inf')

for epoch in range(20):  # 에포크 수
    # 훈련 루프
    bert_symp.train()
    total_train_loss = 0
    for inputs, masks, labels in train_loader:
        inputs, masks, labels = inputs.to(device), masks.to(device), labels.to(device)

        optimizer.zero_grad()
        logits, mean_uncertainty = bert_symp(inputs, masks)
        loss = loss_func(logits, labels.float())
        loss.backward()
        optimizer.step()
        scheduler.step()
        total_train_loss += loss.item()

    avg_train_loss = total_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # 검증 루프
    bert_symp.eval()
    total_val_loss = 0
    with torch.no_grad():
        for inputs, masks, labels in val_loader:
            inputs, masks, labels = inputs.to(device), masks.to(device), labels.to(device)
            logits, mean_uncertainty = bert_symp(inputs, masks)
            loss = loss_func(logits, labels.float())
            total_val_loss += loss.item()

    avg_val_loss = total_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)

    print(f"Epoch {epoch+1}, Training Loss: {avg_train_loss}, Validation Loss: {avg_val_loss}")

    # Early Stopping 기준 검증
    if avg_val_loss < last_loss:
        last_loss = avg_val_loss
        early_stop_counter = 0
    else:
        early_stop_counter += 1
        if early_stop_counter >= 3:
            print("Early stopping...")
            break
```

```python
# 모델 평가
average_threshold = 0.2453  # 평균 확률을 임계값으로 설정

bert_symp.eval()
predictions, true_labels = [], []

for inputs, masks, labels in val_loader:
    inputs, masks, labels = inputs.to(device), masks.to(device), labels.to(device)
    with torch.no_grad():
        logits, mean_uncertainty = bert_symp(inputs, masks)  # 모델 출력

        probabilities = torch.sigmoid(logits)  # 확률값 계산

        # 평균 임계값으로 예측 변환
        predicted_labels = (probabilities > average_threshold).int()
        predictions.append(predicted_labels)
        true_labels.append(labels)

# 평가 지표 계산
# predictions과 true_labels는 (batch_size, num_labels) 형태로, 이를 (num_samples * num_labels,)로 펼쳐야 함
predictions = torch.cat(predictions, dim=0).cpu().numpy()  # (num_samples * num_labels,)
true_labels = torch.cat(true_labels, dim=0).cpu().numpy()  # (num_samples * num_labels,)

# precision, recall, f1 score 계산 (다중 레이블 문제에서는 'samples' 평균이 적합)
precision = precision_score(true_labels, predictions, average='samples', zero_division=1)
recall = recall_score(true_labels, predictions, average='samples', zero_division=1)
f1 = f1_score(true_labels, predictions, average='samples', zero_division=1)

# 결과 출력
print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1 Score: {f1:.4f}")
```

```python
# 입력 데이터
texts = [
    "요즘 우울감이 심해지고 잠도 잘 못자요.",
    "불안감이 지속되고 가끔 숨이 잘 안 쉬어져요."
]

# 입력 데이터 토큰화 및 전처리
encodings = tokenizer(
    texts,
    padding=True,
    truncation=True,
    max_length=128,
    return_tensors="pt"
)

input_ids = encodings['input_ids'].to(device)
attention_mask = encodings['attention_mask'].to(device)

# 모델 예측
bert_symp.eval()  # 평가 모드로 전환
with torch.no_grad():
    logits, _ = bert_symp(input_ids, attention_mask)  # 로짓과 불필요한 다른 값 출력
    probabilities = torch.sigmoid(logits)  # 확률값 계산

# 임계값 적용
threshold = 0.2453  # 평가 단계에서 사용한 평균 임계값
predicted_labels = (probabilities > threshold).int()

# 레이블 디코딩 (레이블 이름 매핑)
decoded_predictions = []
for probs in predicted_labels:
    decoded = [target_classes[i] for i, p in enumerate(probs) if p == 1]
    decoded_predictions.append(decoded)
```

```python
# 결과 출력
for text, prediction in zip(texts, decoded_predictions):
    print(f"Input Text: {text}")
    print(f"Predicted Labels: {prediction}")
    print("-" * 50)
```
