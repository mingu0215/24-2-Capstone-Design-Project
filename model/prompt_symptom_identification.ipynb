```python
from tqdm import tqdm

def split_text(text, prefix='Sentence_'):
    parts = []
    start = 0
    while True:
        start_loc = text.find(prefix, start)
        if start_loc == -1:
            break  # 더 이상 prefix를 찾을 수 없으면 반복 종료
        end_loc = text.find(prefix, start_loc + 1)
        if end_loc == -1:
            parts.append(text[start_loc:])  # 마지막 세그먼트 추가
            break
        parts.append(text[start_loc:end_loc])  # 다음 prefix 시작 지점까지의 텍스트를 추가
        start = end_loc

    return parts

# 텍스트를 분할합니다.
parts = split_text(user_content)

# 각 부분을 모델에 전송하여 결과를 얻습니다.
symptoms_results = []
for part in tqdm(parts):
    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "system",
             "content": symptoms_prompt},
            {"role": "user",
             "content": part}
        ],
        temperature=0.4
    )

    # 각 질환에 대한 라벨을 간결하게 추출하여 저장
    for line in completion.choices[0].message.content.strip().split('\n'):
        if line.startswith('Sentence_'):
            affects = line.split(':')[-1].strip()
            symptoms_results.append(f"{line.split(':')[0]}: {affects}")

# 결과를 출력
for result in symptoms_results:
    print(result)
```

```python
symptoms_list = []

# symptoms_results에서 인덱스와 증상 파싱
for result in symptoms_results:
    index, symptoms = result.split(':', 1)  # ':' 기준으로 인덱스와 증상을 분리
    index = int(index.split('_')[1])  # "Sentence_XXXX"에서 숫자만 추출
    symptoms = symptoms.strip().split('\\')  # '\\'로 나누어 리스트로 변환
    symptoms_list.append((index, symptoms))  # 인덱스와 증상을 튜플로 저장

# 데이터프레임에 맞는 길이로 초기화
df['Symptoms_gpt'] = None

# 인덱스에 맞춰 symptoms 추가
for index, symptoms in symptoms_list:
    if index in df.index:
        df.at[index, 'Symptoms_gpt'] = symptoms

# 결측값 있는 행 제거
df.dropna(subset=['Symptoms_gpt'], inplace=True)

# 결과 출력
print(df['Symptoms_gpt'])
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/성균관대학교 과제/3학년 2학기/캡스톤설계프로젝트/symptom_identification.csv', encoding = 'utf-8-sig')

df.head()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.nn import functional as F
from transformers import AutoModel, AutoTokenizer
from torch.cuda.amp import autocast  # 반정밀도 계산을 위한 자동 혼합 정밀도
import math

# SpectralNorm 클래스
class SpectralNorm(nn.Module):
    def __init__(self, module, name='weight', power_iterations=1, eps=1e-12):
        super(SpectralNorm, self).__init__()
        self.module = module
        self.name = name
        self.power_iterations = power_iterations
        self.eps = eps
        self.register_buffer('u', torch.randn(1, module.weight.size(0)))
        self.register_buffer('v', torch.randn(1, module.weight.size(1)))

    def forward(self, *args):
        weight = self.module.weight
        for _ in range(self.power_iterations):
            self.u = F.normalize(torch.matmul(self.u, weight), dim=-1, eps=self.eps)
            self.v = F.normalize(torch.matmul(self.v, weight.T), dim=-1, eps=self.eps)
        sigma = torch.matmul(self.u, torch.matmul(weight, self.v.T))
        weight = weight / sigma
        return self.module(*args)

# SNGP 관련 클래스 및 함수들
def BertLinear(i_dim, o_dim, bias=True):
    m = nn.Linear(i_dim, o_dim, bias)
    nn.init.normal_(m.weight, std=0.02)
    if bias:
        nn.init.constant_(m.bias, 0.)
    return m

def RandomFeatureLinear(i_dim, o_dim, bias=True, require_grad=False):
    m = nn.Linear(i_dim, o_dim, bias)
    nn.init.normal_(m.weight, mean=0.0, std=0.05)
    m.weight.requires_grad = require_grad
    if bias:
        nn.init.uniform_(m.bias, a=0.0, b=2. * math.pi)
        m.bias.requires_grad = require_grad
    return m

class SNGP(nn.Module):
    def __init__(self, backbone, hidden_size=256, num_labels=48, gp_kernel_scale=2.0, num_inducing=128,
                 gp_output_bias=0., n_power_iterations=1, spec_norm_bound=0.95, device='cuda'):
        super(SNGP, self).__init__()
        self.backbone = backbone
        self.num_labels = num_labels
        self.hidden_size = hidden_size
        self.gp_kernel_scale = gp_kernel_scale
        self.num_inducing = num_inducing

        # BERT의 마지막 레이어
        self.last_pooled_layer = SpectralNorm(BertLinear(hidden_size, hidden_size),
                                               power_iterations=n_power_iterations)

        self.gp_input_scale = 1. / math.sqrt(gp_kernel_scale)
        self._random_feature = RandomFeatureLinear(hidden_size, num_inducing)
        self._gp_output_layer = nn.Linear(num_inducing, num_labels, bias=False)
        self._gp_output_bias = torch.tensor([gp_output_bias] * num_labels).to(device)

        self._gp_input_normalize_layer = torch.nn.LayerNorm(hidden_size)
        self.precision_matrix = torch.nn.Parameter(torch.eye(num_inducing).to(device), requires_grad=False)

    def gp_layer(self, gp_inputs):
        # gp_inputs를 정규화
        gp_inputs = self._gp_input_normalize_layer(gp_inputs)

        # 랜덤 특징 변환
        gp_feature = self._random_feature(gp_inputs)

        # cos() 함수로 변환 후 스케일링
        gp_feature = torch.cos(gp_feature) * self.gp_input_scale
        gp_output = self._gp_output_layer(gp_feature) + self._gp_output_bias

        return gp_feature, gp_output  # 두 값을 모두 반환

    def forward(self, input_ids, attention_mask):
        # BERT 모델에서 나온 출력
        outputs = self.backbone(input_ids=input_ids, attention_mask=attention_mask)

        # 'last_hidden_state'에서 첫 번째 토큰(CLS)을 추출
        pooled_output = outputs.last_hidden_state[:, 0, :]  # CLS 토큰 추출 (배치 크기, hidden_size)

        # SNGP 레이어에 입력
        gp_feature, logits = self.gp_layer(pooled_output)

        return logits  # logits만 반환


# BERT + SNGP 결합 모델
class BertMultiLabelClassificationWithSNGP(nn.Module):
    def __init__(self, model_name, num_labels, dropout_ratio=0.2, num_samples=10):
        super(BertMultiLabelClassificationWithSNGP, self).__init__()
        self.bert = AutoModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(dropout_ratio)
        self.num_samples = num_samples
        self.sngp = SNGP(self.bert, hidden_size=self.bert.config.hidden_size, num_labels=num_labels)
        # self.fc의 입력 차원을 96에서 49로 수정
        self.fc = nn.Linear(96, num_labels)

    def forward(self, input_ids, attention_mask, symptom_uncertainty=None):
        self.train()
        predictions, uncertainties = [], []

        for _ in range(self.num_samples):
            with autocast():  # 모델의 계산을 반정밀도에서 진행
                bert_outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
                pooled_output = bert_outputs.last_hidden_state[:, 0, :]
                pooled_output = self.dropout(pooled_output)

                gp_feature, gp_output = self.sngp.gp_layer(pooled_output)
                probabilities = torch.sigmoid(gp_output)

                predictions.append(probabilities)
                uncertainties.append(torch.std(probabilities, dim=0))

        mean_predictions = torch.mean(torch.stack(predictions), dim=0)  # [32, 48]
        mean_uncertainty = torch.mean(torch.stack(uncertainties), dim=0)  # [48]

        # mean_uncertainty를 [32, 1]로 확장
        mean_uncertainty = mean_uncertainty.unsqueeze(0)  # [1, 48]
        mean_uncertainty = mean_uncertainty.expand(mean_predictions.size(0), -1)  # [32, 48]으로 확장

        # symptom_uncertainty가 None이 아닐 때 combined_uncertainty를 만듦
        if symptom_uncertainty is not None:
            combined_uncertainty = torch.cat([symptom_uncertainty.unsqueeze(1), mean_uncertainty], dim=1)  # [32, 49] 형태
        else:
            combined_uncertainty = mean_uncertainty  # [32, 48]

        # mean_predictions와 combined_uncertainty를 결합
        combined_input = torch.cat([mean_predictions, combined_uncertainty], dim=1)  # [32, 96]로 결합

        # fc 계층을 통과
        logits = self.fc(combined_input)  # [32, 49] 형태로 처리됨

        return logits, mean_uncertainty  # logits과 mean_uncertainty 반환
```

```python
# 모델 및 토크나이저 로드
# 모델 초기화
model = BertMultiLabelClassificationWithSNGP('klue/bert-base', num_labels=48, dropout_ratio=0.2)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')  # GPU가 사용 가능하면 GPU로 설정
model.to(device)  # 모델을 GPU 또는 CPU로 이동

# 토크나이저 초기화
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

## SNGP 값 추출 함수
def extract_sngp_values(df, model, tokenizer, device, batch_size=16):
    sngp_values = []
    model.eval()

    with torch.no_grad():
        for i in range(0, len(df), batch_size):
            # 배치 데이터 추출
            batch_texts = df['Symptoms_gpt'][i:i + batch_size].tolist()

            # 토크나이즈 및 배치 처리
            inputs = tokenizer(
                batch_texts,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=128
            )

            # 입력 데이터를 모델과 동일한 디바이스로 이동
            inputs = {key: value.to(device) for key, value in inputs.items()}

            # 모델로 예측 수행
            outputs = model(
                input_ids=inputs['input_ids'],
                attention_mask=inputs['attention_mask']
            )

            # SNGP 불확실성 값 추출
            logits, uncertainties = outputs

            # 디버깅: logits 및 uncertainties 확인
            print(f"logits: {logits}")
            print(f"uncertainties: {uncertainties}")

            # 결과 저장
            sngp_values.extend(uncertainties.tolist())  # 각 배치의 결과를 리스트에 추가

    return sngp_values

# 결측값 처리
df = df.dropna(subset=['Symptoms_gpt'])

# SNGP 값 추출
df['SNGP_values'] = extract_sngp_values(df, model, tokenizer, device)

print(df['SNGP_values'])
```

```python
import numpy as np
import pandas as pd

# 각 증상별 SNGP 평균 계산 함수
def calculate_sngp_per_symptom(df):
    # SNGP 값을 데이터프레임으로 변환 (48개 증상이 각 열에 해당)
    sngp_matrix = np.array(df['SNGP_values'].tolist())  # SNGP 값 리스트를 NumPy 배열로 변환

    # 각 증상(열)의 평균 계산
    symptom_means = np.mean(sngp_matrix, axis=0)

    return symptom_means

# 전체 평균 계산 (각 증상의 평균)
symptom_means = calculate_sngp_per_symptom(df)

# 결과 출력 (요구한 형식으로 변환)
formatted_means = [round(mean, 4) for mean in symptom_means]
print(formatted_means)
```
